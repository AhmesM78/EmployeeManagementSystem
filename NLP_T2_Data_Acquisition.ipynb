```python
from youtube_comment_downloader import YoutubeCommentDownloader
import pandas as pd

downloader = YoutubeCommentDownloader()

url = "https://www.youtube.com/watch?v=J5HzbEc_atI"

comments = []
for i, comment in enumerate(downloader.get_comments_from_url(url)):
    comments.append({
        "username": comment['author'],
        "comment": comment['text'],
        "likes": comment['votes'],
        "time": comment['time']
    })
    if i >= 1000:
        break

df = pd.DataFrame(comments)
df.to_csv("youtube_comments.csv", index=False)

df.head()
```

```python
from google.colab import sheets
sheet = sheets.InteractiveSheet(df=df)
```

```python
from google_play_scraper import reviews, Sort
import pandas as pd

result, _ = reviews(
    'com.mobilechess.gp',
    lang='id',
    country='id',
    sort=Sort.NEWEST,
    count=1000
)

data = []
for r in result:
    data.append({
        "username": r['userName'],
        "rating": r['score'],
        "review": r['content'],
        "date": r['at']
    })

df = pd.DataFrame(data)
df.to_csv("playstore_reviews.csv", index=False)

df.head()
```

```python
from google.colab import sheets
sheet = sheets.InteractiveSheet(df=df)
```
